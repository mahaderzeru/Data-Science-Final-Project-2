### Analysis

Research Question: How did the yearly compensation for tech jobs at Amazon, Microsoft, and Google in the US change during the COVID-19 pandemic than in the years before? 

- identify patterns in yearly compensation, factor in job title and years of experience to get more specific

```
 #Splitting the data set into pre-pandemic and pandemic years
pre_pandemic_years = [2018, 2019]
pandemic_years = [2020, 2021]
companies = ['Google', 'Amazon', 'Microsoft']

pre_pandemic_data = df1[df1['year'].isin(pre_pandemic_years)]
pandemic_data = df1[df1['year'].isin(pandemic_years)]
```

```
# Features
Y = df1['total_yearly_compensation'].astype(float)
```

```
# Turning categorical data to numerical data
X = pd.get_dummies(df1[['year', 'title', 'years_of_experience']], columns=['title'])
```

```
# Splitting the data set into testing and training sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```
# Linear Regression model

# Iterate through each company in the list of companies
for company in companies: 
    pre_pandemic_company_data = pre_pandemic_data[pre_pandemic_data['company'] == company]
    pandemic_company_data = pandemic_data[pandemic_data['company'] == company]

    #concatenate the data for the company
    company_data = pd.concat([pre_pandemic_company_data, pandemic_company_data])

    model = LinearRegression()

    model.fit(X_train, Y_train)
    predictions = model.predict(X_test)

    print(f'\nCompany: {company}')
    print(mean_squared_error(Y_test, predictions))
    print(r2_score(Y_test, predictions))
```

```
#DecisionTree Regression model
for company in companies: 
    pre_pandemic_company_data = pre_pandemic_data[pre_pandemic_data['company'] == company]
    pandemic_company_data = pandemic_data[pandemic_data['company'] == company]

    #concatenate the data for the company
    company_data = pd.concat([pre_pandemic_company_data, pandemic_company_data])

    model = DecisionTreeRegressor(max_depth = 5, min_samples_split = 12, min_samples_leaf = 6)

    model.fit(X_train, Y_train)
    predictions = model.predict(X_test)

    print(f'\nCompany: {company}')
    print(mean_squared_error(Y_test, predictions))
    print(r2_score(Y_test, predictions))
```

```
#GradientBoosting Regression model
for company in companies: 
    pre_pandemic_company_data = pre_pandemic_data[pre_pandemic_data['company'] == company]
    pandemic_company_data = pandemic_data[pandemic_data['company'] == company]

    #concatenate the data for the company
    company_data = pd.concat([pre_pandemic_company_data, pandemic_company_data])

    model = GradientBoostingRegressor(n_estimators = 100, learning_rate = 0.1, max_depth = 5)

    model.fit(X_train, Y_train)
    predictions = model.predict(X_test)

    print(f'\nCompany: {company}')
    print(mean_squared_error(Y_test, predictions))
    print(r2_score(Y_test, predictions))
```

```
for company in companies:
    # Filter data for the current company
    pre_pandemic_company_data = pre_pandemic_data[pre_pandemic_data['company'] == company]
    pandemic_company_data = pandemic_data[pandemic_data['company'] == company]

    # Concatenate the data for the company
    company_data = pd.concat([pre_pandemic_company_data, pandemic_company_data])

    # Fit OLS regression model
    X_company = sm.add_constant(company_data.drop(columns=['total_yearly_compensation']))
    model = sm.OLS(company_data['total_yearly_compensation'], X_company)
    results = model.fit()

    # Print company name and summary of regression results
    print(f'\nCompany: {company}')
    print(results.summary())

    # Make predictions
    X_test_company = sm.add_constant(X_test)
    predictions = results.predict(X_test_company)

    # Print evaluation metrics
    print("Mean Squared Error:", mean_squared_error(Y_test, predictions))
    print("R-squared:", r2_score(Y_test, predictions))

    # Visualization
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=Y_test, y=predictions)
    plt.xlabel("Actual Total Yearly Compensation")
    plt.ylabel("Predicted Total Yearly Compensation")
    plt.title(f"Actual vs Predicted Total Yearly Compensation for {company}")
    plt.show()
```
